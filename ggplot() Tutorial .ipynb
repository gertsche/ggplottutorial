```python
(ggplot(mpg)         # defining what data to use
 + aes(x='class')    # defining what variable to use
 + geom_bar(size=20) # defining the type of plot to use
)

# in R, indentation is important to good style
# need to use parentheses in python because it doesn't recognize code spilling over onto multiple lines
```

```python
(ggplot(mpg)
 + aes(x='class')
 + geom_bar(size=20)
 + coord_flip()        # flipping the x- and y-axes
 + labs(title='Number of Vehicles per Class', x='Vehicle Class', y='Number of Vehicles') # customizing labels
)

# adding more customization
```

```python
(ggplot(mtcars) 
    + aes(x='wt',y='mpg')
    + geom_point()
    + xlab('Weight (1000s of lbs)') 
    + ylab('Miles per Gallon') 
    + geom_smooth(colour = 'red')
)

# another example, this time using a scatterplot
```

```python
(ggplot(diamonds) 
    + aes(x = 'carat', y='..density..')
    + geom_histogram(bins = 10, fill = "lightblue", color = "black")
    + xlab('Number of Carats')
)
```

```python
(ggplot(mtcars)
    + aes(x='mpg', fill='cyl') 
    + geom_histogram(position="identity", fill = 'darkblue', colour="blue", alpha=0.2, bins = 10) 
)
```

```python
# example with facet_wrap
(ggplot(diamonds)
    + aes(x='price', color='clarity')
    + geom_density() 
    + scale_color_brewer(type='div', palette=7) 
    + facet_wrap('cut')
)
```
